### <font color=ligtblue><b> üíª Kompletn√≠ k√≥d:</b>

````python
# Zadan√© hodnoty
kosik = dict()
oddelovac = "=" * 40
sklad = {
    "mleko": [30, 5],  # index '0'-> cena, index '1' -> pocet
    "maso": [100, 1],
    "banan": [30, 10],
    "jogurt": [10, 5],
    "chleb": [20, 5],
    "jablko": [10, 10],
    "pomeranc": [15, 10]
}
nabidka = """
+-----------+----------+
| POTRAVINA |   CENA   |
+-----------+----------+
| mleko     |    30,-  |
| maso      |   100,-  |
| banan     |    30,-  |
| jogurt    |    10,-  |
| chleb     |    20,-  |
| jablko    |    10,-  |
| pomenrac  |    15,-  |
+-----------+----------+
"""

# √övodn√≠ sekce s uv√≠t√°n√≠m a v√Ωpisem zbo≈æ√≠
print(
    "Vitejte u naseho nakupniho kosiku!".center(len(oddelovac)),
    oddelovac,
    nabidka,
    sep="\n"
)

# Cyklus, kter√Ω p≈ôid√°v√° zbo≈æ√≠ do ko≈°√≠ku tak dlouho, jak u≈æivatel zam√Ω≈°l√≠
#while (zbozi := input("Zbozi:")) != "q": # walrus nefunkcni v Colabu
prepinac = True

while prepinac:
    select = (input('Vlozit do kosiku: '))
    if select == 'q':
        prepinac = False

    elif select not in sklad:
        print(f'{select} neni v nabidce')

    elif select not in kosik and sklad[select][1] > 0:
        kosik[select] = [sklad[select][0], 1]
        sklad[select][1] -= 1

    elif select in kosik and sklad[select][1] > 0:
        kosik[select][1] += 1
        sklad[select][1] -= 1

    elif sklad[select][1] == 0:
        print('Toto zbozi jiz neni skladem')

# Fin√°ln√≠ v√Ωpis ko≈°√≠ku po n√°kupu
else:
    print(oddelovac, kosik, sep="\n")


````
