## **Smyƒçka WHILE**

---

````python
podminka = 1
````

````python
while podminka < 10:
    print(f'V tomto cyklu je {podminka=}, coz je mensi nez 10.')
    podminka = podminka + 1

print(f'Ted uz ale {podminka=} neni mensi nez 10')

print('Kod pokracuje pod smyckou.')
````

<br>


* `while` je kl√≠ƒçov√Ω v√Ωraz v z√°hlav√≠ (z√°pisu) cyklu,
* `podminka < 10` je podm√≠nka. Pokud je vyhodnocen√° jako `True`, proveƒè dal≈°√≠ krok; pokud `False`, pokraƒçuj se z√°pisem pod

## **WHILE loop s dopl≈àuj√≠c√≠ podm√≠nkou**

---

Pokud je pot≈ôeba dopsat je≈°tƒõ dopl≈àuj√≠c√≠ podm√≠nku, um√≠ i `while` pracovat s tƒõmi dopl≈àky (stejnƒõ jako cyklus `for`):

````python
index = 0
````

````python
while index < 20:
    if len(str(index)) != 2:
        index = index + 1
        continue
    print(f"index={index}")
    index += 1
````

*Pozn√°mka:* Opatrnƒõ, je rozumn√© zkusit se nad z√°pisem zamyslet, jestli neexistuje mo≈ænost, jak podm√≠nku upravit p≈ô√≠mo v p≈ôedpisu

````python
index = 10
````

````python
while index < 20 and len(str(index)) == 2:
    print(f"index={index}")
    index += 1
````

Nekoneƒçn√Ω cyklus s pomoc√≠ `while` je samoz≈ôejmƒõ koncept, kter√©ho se bƒõ≈ænƒõ vyu≈æ√≠v√°, pokud je podm√≠nka a pr≈Øbƒõh validn√≠:

#### **1. Poƒçet v√Ωskyt≈Ø**

Metoda `.get` je velmi u≈æiteƒçn√°, kdy≈æ pot≈ôebujeme spoƒç√≠tat poƒçet v√Ωskyt≈Ø prvk≈Ø. Abychom pochopili n√°sleduj√≠c√≠ k√≥d, mus√≠me vƒõdƒõt nƒõco o smyƒçce while.

M≈Ø≈æeme postupovat t√≠mto zp≈Øsobem:

````python
barvy = [
    'zelena', 'modra', 'cerna', 'cervena',
    'cervena', 'zluta', 'modra', 'seda',
    'cerna' , 'cervena', 'zelena'
    ]

pocet_barev = {}

while barvy:
    barva = barvy.pop()

    if barva not in pocet_barev:
        pocet_barev[barva] = 0

    pocet_barev[barva] = pocet_barev[barva] + 1
````

````python
pocet_barev
````

Efektivnƒõj≈°√≠ je ale tento postup:

````python
barvy = [
    'zelena', 'modra', 'cerna', 'cervena',
    'cervena', 'zluta', 'modra', 'seda',
    'cerna' , 'cervena', 'zelena'
    ]

pocet_barev = {}

while barvy:
    barva = barvy.pop()
    pocet_barev[barva] = pocet_barev.get(barva,0) + 1
````

````python
mesta = ['Praha', 'Brno', 'Ostrava', 'Plzen', 'Liberec']
while mesta:
    if len(mesta) < 3:
        break
    print(mesta.pop())
print('Hotovo.')
````

Tento k√≥d je stejn√Ω jako v p≈ôedchoz√≠m p≈ô√≠padu, ale m√≠sto `break` je zde ohl√°≈°en√≠ `continue`.

Jak√° bude hodnota promƒõnn√© `vysledek`, kdy≈æ spust√≠me n√°sleduj√≠c√≠ k√≥d?

````python
vysledek = ""

n = 5
while n > 0:
    n -= 1
    if (n % 2) == 0:
        continue

    mesta = ['Praha', 'Brno', 'Ostrava']
    while mesta:
        vysledek += str(n) + mesta.pop(0)
        if len(mesta) < 2:
            break
````
