````python
# Multi_modul.py

muj_string = "Python je cool!"

moje_cislo = 39

seznam = [100, 300, 10000]

slovnicek = {
    'jmeno': 'Radim',
    'prijmeni': 'Jedlicka'
}


print(muj_string)
print(moje_cislo)
print(seznam)
print(slovnicek)
````

### **Pomocn√° klauzule `if __name__ == '__main__':`**


Pokud bude≈° cht√≠t uchovat oboj√≠ funkcionalitu, potom mus√≠≈° napsat podm√≠nkov√Ω z√°pis s `if`, do jeho≈æ odsazen√©ho z√°pisu vlo≈æ√≠≈° v≈°echno, co bude≈° cht√≠t spou≈°tƒõt jako skript:

[V√≠ce o t√©to klauzuli si m≈Ø≈æete p≈ôeƒç√≠st zde](https://realpython.com/if-name-main-python/)

````python
# Multi_modul.py

muj_string = "Python je cool!"

moje_cislo = 39

seznam = [100, 300, 10000]

slovnicek = {
    'jmeno': 'Radim',
    'prijmeni': 'Jedlicka'
}


# tuto podm√≠nkovou vƒõtev nyn√≠ p≈ôi nahr√°v√°n√≠ knihovny nespust√≠≈°
if __name__ == '__main__':
    print(muj_string)
    print(moje_cislo)
    print(seznam)
    print(slovnicek)
````
